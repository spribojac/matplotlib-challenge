### Analysis

- Regarding the boxplots it can be suggested that the Capomulin drug regimen was comparable to the Ramicane drug regimen in terms of reduction in tumor size, but both were more effective than Infubinol and Ceftamin.

- Looking at mouse weigth vs average tumor size it is noticeable that average tumour size increases with weight of mouse. This may suggest that drug treatment is less effective the heavier the mouse.

- Building on the previous point, the correlation between both mouse weight and average tumor size is 0.84 which is indicative of a strong positive correlation between these two factors.

- There is only one outlier in the 4 drug regimens investigated (Infubinol) which suggests that the data is normally distributed and, therefore reliable conclusions can be made from this data.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import sem
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_data_df = pd.merge(study_results, mouse_metadata, on='Mouse ID', how='outer')

# Display the data table for preview
mouse_data_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Set the regimen as the index
drug_index_df = cleaned_df.set_index('Drug Regimen')

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean_tumor_size = drug_index_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].mean()
median_tumor_size = drug_index_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].median()
std_tumor_size = drug_index_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].std()
var_tumor_size = drug_index_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].var()
sem_tumour_size = drug_index_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].sem
# Assemble the resulting series into a single summary DataFrame.
summary_statistics_df = pd.DataFrame({
    'Mean Tumor Size': mean_tumor_size,
    'Median Tumor Size': median_tumor_size,
    'Variance of Tumor Size': var_tumor_size,
    'Standard Deviation of Tumor Size': std_tumor_size,
    'Standard Errors of Tumor Size': std_tumor_size,
})

summary_statistics_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
summary_df = cleaned_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(
    Mean_Tumor_Size='mean',
    Median_Tumor_Size='median',
    Variance_of_Tumor_Size='var',
    Std_Deviation_of_Tumor_Size='std',
    SEM_of_Tumor_Size='sem'
).reset_index()

# Display the summary statistics table
print(summary_statistics_df)
# Using the aggregation method, produce the same summary statistics in a single line
new_df = pd.DataFrame(summary_statistics_df)
new_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

drugs = ['Capomulin', 'Ramicane', 'Infubinol','Ceftamin']

# Step 1: Get the last (greatest) timepoint for each mouse
max_tp_df = cleaned_df.groupby('Mouse ID')['Timepoint'].max().reset_index()
max_tp_df = max_tp_df.rename(columns={'Timepoint': 'Max Timepoint'})

# Step 2: Merge the maximum timepoint DataFrame with the original DataFrame
final_tp_df = cleaned_df.merge(max_tp_df, on='Mouse ID', how='inner')

# Filter the DataFrame to include only the specified treatment regimens
final_tp_df = final_tp_df[final_tp_df['Drug Regimen'].isin(drugs)]

final_tp_df.reset_index(drop=True, inplace=True)  # Resetting the index for better display
final_tp_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ['Capomulin', 'Ramicane', 'Infubinol','Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in drugs:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    drug_tp_df = final_tp_df.loc[final_tp_df['Drug Regimen'] == drug]
    
    # Add subset: Get the tumor volumes for the last timepoint
    final_volume_df = drug_tp_df.loc[drug_tp_df['Timepoint'] == drug_tp_df['Max Timepoint']]
    
    # Calculate the quartiles and IQR
    quartiles = final_volume_df['Tumor Volume (mm3)'].quantile([0.25, 0.5, 0.75])
    lower_quartile = quartiles[0.25]
    median = quartiles[0.5]
    upper_quartile = quartiles[0.75]
    iqr = upper_quartile - lower_quartile
    
    # Determine the lower and upper bounds for potential outliers
    lower_bound = lower_quartile - 1.5 * iqr
    upper_bound = upper_quartile + 1.5 * iqr
    
    # Identify potential outliers
    outliers = final_volume_df.loc[
        (final_volume_df['Tumor Volume (mm3)'] < lower_bound) | (final_volume_df['Tumor Volume (mm3)'] > upper_bound)]
    
    # Append the tumor volume data to the list for plotting
    tumor_volume.append(final_volume_df['Tumor Volume (mm3)'])
    
    # Quantitatively check for outliers
    outliers_count = (final_volume_df.loc[(final_volume_df['Tumor Volume (mm3)'] >= upper_bound) | 
                                        (final_volume_df['Tumor Volume (mm3)'] <= lower_bound)]).count()
    
    outliers_count = outliers.shape[0]
        
    print(f'IQR for {drug}: {iqr}')
    print(f'Lower Bound for {drug}: {lower_bound}')
    print(f'Upper Bound for {drug}: {upper_bound}')
    print(f'Number of {drug} outliers: {outliers_count}\n')
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
tumor_volume_data = [tumor_volume[0], tumor_volume[1], tumor_volume[2], tumor_volume[3]]

# Create a boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(tumor_volume_data, labels=drugs, showmeans=True)

# Add labels and title
plt.xlabel('Drug Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.title('Final Tumor Volume Distribution for Each Drug Regimen')

# Show the plot
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
# Mouse ID l509

mouse_l509 = cleaned_df.loc[cleaned_df['Mouse ID'] == 'l509']
mouse_l509
plt.plot(mouse_l509['Timepoint'], mouse_l509['Tumor Volume (mm3)'], marker='o')
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Tumor Volume vs Timepoint for Mouse l509 treated with Capomulin')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# Filter by the Capomulin data
capomulin_data = cleaned_df[cleaned_df['Drug Regimen'] == 'Capomulin']

# Group by the weight and average tumor volume and reset the index
average_tumor_size = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
mouse_weights = capomulin_data.groupby('Mouse ID')['Weight (g)'].first()  # Get the weight of the first record for each mouse

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(mouse_weights, average_tumor_size, marker='o', edgecolors='blue')

# Add labels and title
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')

# Show the plot
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
x_values = mouse_weights
y_values = average_tumor_size
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0,50),fontsize=15,color="red")
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
print(f"The r-squared is: {rvalue**2}")
plt.show()

correlation = st.pearsonr(mouse_weights,average_tumor_size)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```
